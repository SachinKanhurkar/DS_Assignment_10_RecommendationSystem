### Problem statement.

Build a recommender system by using cosine simillarties score.

```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('book.csv',encoding='latin-1',index_col=0)
```

```python
data.head()
```

```python
data.rename(columns={'User.ID': 'UserID', 'Book.Title': 'BookTitle', 'Book.Rating':'BookRating'}, inplace=True)
```

```python
data.head()
```

```python
 from pandas.core.reshape.pivot import pivot
```

```python
data.columns
```

```python
data1=data.drop_duplicates(subset="UserID")
```

```python
data1.shape
```

```python
user_data_df =data1.pivot(index='UserID',
                                 columns='BookTitle',
                                 values='BookRating').reset_index(drop=True)
```

```python
user_data_df
```

```python
user_data_df.index = data1['UserID'].unique()
```

```python
user_data_df.head()
```

```python
#Impute those NaNs with 0 values
user_data_df.fillna(0, inplace=True)
```

```python
user_data_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_data_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index =data.UserID.unique()
user_sim_df.columns = data.UserID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
